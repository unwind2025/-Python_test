## Dataset

实际上我们从chapter_2 开始就在使用该模块

```python
from torch.utils.data import Dataset

class RandomDataset(Dataset):
    def __init__(self, labels):
        self.labels = labels
        
    def __getitem__(self, index):
        data = torch.mul(torch.randn(2), 0.1) + self.labels[index]
        label = self.labels[index]
        return data, label
    
    def __len__(self):
        return len(self.labels)
    

rd = RandomDataset(np.random.randint(2, size=1000))
print("数据集的长度：", len(rd))

# 可视化随机数据集
xx = []
yy = []
labels = []

for (x, y), label in rd:
    xx.append(x)
    yy.append(y)
    labels.append(label)
    
    
plt.scatter(xx, yy, s=10, c=labels)
plt.colorbar()
plt.show()
```

## DataLoader

```python
from torch.utils.data import DataLoader

ld = DataLoader(rd, batch_size=8, shuffle=True)

for data, label in ld:
    print(data)
    print(label)
    break
```
